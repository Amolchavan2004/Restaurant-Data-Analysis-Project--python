```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
import re
```

```python
df = pd.read_csv('eazydiner_data.csv')
```

```python
df.head()
```

```python
df['AREA'] = df['AREA'].str.replace('\xa0 \| View All\(\d+\) Outlets', '', regex=True)
```

```python
df
```

```python
df['DISHES'] = df['DISHES'].str.replace('\₹\s(\d*)\sfor two approx', '', regex=True)
```

```python
df
```

```python
df['APPROX'] = df['APPROX'].str.replace('\D', '', regex=True)
```

```python
df
```

```python
def extract_discounts(offer):
    return re.findall(r'(\d+)%', str(offer))

df['DISCOUNTS'] = df['DISC'].apply(extract_discounts)
df['DISCOUNTS'].apply(pd.Series)
```

```python
df['DISCOUNTS'] = df['DISCOUNTS'].apply(lambda x: sum(map(int, x)) if isinstance(x, list) else 0)
df['DISCOUNTS'].head()
```

```python
df['NAME'] = df['NAME'].str[:-3]
```

```python
df
```

```python
locations = ['Ahmedabad', 'Mumbai', 'Pune', 'Bengaluru','Hyderabad']
loc = np.repeat(locations, 135)
df['LOCATION'] = loc
```

```python
df.iloc[:137,:]
```

```python
df[['AREA', 'City']] = df['AREA'].str.split(',', n=1, expand=True)

df = df.drop(columns=['City'])
```

```python
df
```

```python
df = df.rename(columns={'AREA':'ADDRESS', 'APPROX':'APPROX_PRICE','LOCATION':'CITY'})
```

```python
final_cols = [
    'NAME',
    'RATING',
    'ADDRESS', 
    'CITY', 
    'DISHES', 
    'APPROX_PRICE', 
    'DISCOUNTS'
]
df = df[final_cols]
```
